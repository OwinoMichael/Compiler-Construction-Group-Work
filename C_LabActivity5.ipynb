# **1. Review what syntax analysis is and in a few words, describe what this phase of compiler construction entails.**

Syntax analysis is the process of analyzing a string of symbols either in natural language or computer languages that satisfies the rules of a formal grammar. It is the second phase of compilation process and is also referred to as syntactic analysis or parsing. Tokens are taken as input and a parse tree is generated as output

During this phase a structure that represents the program being compiled is built up in memory. This structure called an Abstract Syntax Tree or AST accurately represents the program being compiled and is structured in such a way that it can be used to check the syntax of the source code and in typed languages it can be used to perform type checking of the source program. The AST is also used by subsequent phases of compilation to do such things as intermediate code generation, memory allocation and code optimization.

# **2. Using Flex (lexical analyzer) and Bison (yacc), show how one can develop a simple syntax analyzer.**

A lexical analyzer or scanner converts an incoming stream of characters into an outgoing stream of tokens. The parser section of a compiler then translates the input stream of tokens into a series of semantic actions. The code generator, another part of your complete compiler, is isolated from the parser section as much as possible. The parser receives its tokens from the lexical scanner as the return value of yylex(). The yylex() function assembles the tokens by matching patterns against input strings. A particular instance of a token is called a lexeme.

# **Example: Shift Reduce Parser**
Shift reduce parsing is a process of reducing a string to the start symbol of a grammar.
Shift reduce parsing uses a stack to hold the grammar and an input tape to hold the string.
Shift reduce parsing
Sift reduce parsing performs the two actions: shift and reduce. That's why it is known as shift reduces parsing.
At the shift action, the current symbol in the input string is pushed to a stack.
At each reduction, the symbols will replaced by the non-terminals. The symbol is the right side of the production and non-terminal is the left side of the production.
